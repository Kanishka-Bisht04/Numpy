```python
file_path = r'C:\Users\kanis\OneDrive\Pictures\Screenshots\numpy–logo.png'
```

```python
try:
    logo = mpimg.imread(file_path)

    if logo.dtype == np.uint8:
        logo = logo / 255.0
        
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("NumPy Logo")
    plt.grid(False)
    
    dark_logo = logo.copy()
    dark_logo[..., :3] = 1 - dark_logo[..., :3]

        
    
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("NumPy Dark Logo")
    plt.grid(False)

except FileNotFoundError:
    print("NumPy logo file not found")

except Exception as e:
    print("Something else went wrong:", e)
```
